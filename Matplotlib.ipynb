```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
```

```python
x_sqaured = x**2
```

```python
plt.plot(x,x, marker = "*")
plt.plot(x, x_sqaured, marker = "o" )
plt.xlabel("Number")
plt.show()
```
